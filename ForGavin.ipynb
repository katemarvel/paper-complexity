```python
import get_data
import numpy as np
import matplotlib.pyplot as plt
```

```python
sea_ice_data=get_data.data
```

Now, parse the data into arrays of varying degrees of complexity

```python
#Added by kate
colors=[]
very_simple=[]
simple=[]
intermediate=[]
complex_m=[]
for model in range(len(sea_ice_data)):
    ensemble=sea_ice_data[model]
    for member in range(len(ensemble)):
        complexity, mean, trend, series=ensemble[member]
        colors+=[get_data.colors_complexity[complexity-1]]
        if complexity==1:
            very_simple+=[[mean,trend]]
            
        elif complexity==2:
            simple+=[[mean,trend]]
        elif complexity==3:
            intermediate+=[[mean,trend]]
        elif complexity==4:
            complex_m+=[[mean,trend]]
```

Plot the trend vs the mean, with colors determined by the colormap in get_data.py

```python
total = np.array( very_simple+simple+intermediate+complex_m)                 
means = total[:,0]
trends = total[:,1]
plt.scatter(means,trends,color=colors)
plt.xlabel("Mean")
plt.ylabel("Trend")
            
```
